# Import Required Libraries
import json
import os

## ‚öôÔ∏è Configuration Section

This section defines:
- File name for storing student data
- Admin credentials

```python
FILE_NAME = "students.json"

ADMIN_USERNAME = "Rahul Kumar"
ADMIN_PASSWORD = "1234"
```

## üìÇ File Handling Functions

These functions are responsible for:
- Loading data from JSON file
- Saving data to JSON file

```python
def load_data():
    if not os.path.exists(FILE_NAME):
        return []
    with open(FILE_NAME, "r") as f:
        return json.load(f)
```

```python
def save_data(data):
    with open(FILE_NAME, "w") as f:
        json.dump(data, f, indent=4)
```

## üîê Login System

This section includes:
- Admin Login
- Student Login

```python
def admin_login():
    print("\n--- Admin Login ---")
    u = input("Username: ")
    p = input("Password: ")

    if u == ADMIN_USERNAME and p == ADMIN_PASSWORD:
        print("Admin Login Successful\n")
        admin_menu()
    else:
        print("Invalid Admin Credentials")


def student_login():
    students = load_data()
    roll = input("Enter Roll No: ")
    password = input("Enter Password: ")

    for s in students:
        if s["roll"] == roll and s["password"] == password:
            print("Student Login Successful\n")
            student_menu(s)
            return
    print("Invalid Roll No or Password")
```

## üõ†Ô∏è Admin Functionalities

Admin can:
- Add Student
- View Students
- Update Student
- Delete Student

```python
def add_student():
    students = load_data()

    roll = input("Roll No: ")

    for s in students:
        if s["roll"] == roll:
            print("Roll Number Already Exists!")
            return

    name = input("Name: ")
    age = input("Age: ")
    course = input("Course: ")
    password = input("Set Password: ")

    marks = {
        "math": int(input("Math Marks: ")),
        "science": int(input("Science Marks: ")),
        "english": int(input("English Marks: "))
    }

    student = {
        "roll": roll,
        "name": name,
        "age": age,
        "course": course,
        "password": password,
        "marks": marks
    }

    students.append(student)
    save_data(students)
    print("Student Added Successfully ‚úÖ\n")


def view_students():
    students = load_data()

    if not students:
        print("No Students Found")
        return

    print("Total Students:", len(students))
    print("-------------------------")

    for s in students:
        print("Roll:", s["roll"])
        print("Name:", s["name"])
        print("Course:", s["course"])
        print("-------------------------")


def update_student():
    students = load_data()
    roll = input("Enter Roll No to Update: ")

    for s in students:
        if s["roll"] == roll:
            s["name"] = input("New Name: ")
            s["age"] = input("New Age: ")
            s["course"] = input("New Course: ")

            save_data(students)
            print("Student Updated Successfully ‚úÖ\n")
            return

    print("Student Not Found ‚ùå")


def delete_student():
    students = load_data()
    roll = input("Roll No to Delete: ")

    new_data = [s for s in students if s["roll"] != roll]

    if len(new_data) == len(students):
        print("Student Not Found ‚ùå")
    else:
        save_data(new_data)
        print("Student Deleted Successfully ‚úÖ\n")
```

## üìä Student Dashboard

Students can:
- View Profile
- View Marks
- See Total & Grade

```python
def student_menu(student):
    marks = student["marks"]
    total = sum(marks.values())
    percentage = total / 3

    if percentage >= 75:
        grade = "A"
    elif percentage >= 60:
        grade = "B"
    elif percentage >= 40:
        grade = "C"
    else:
        grade = "F"

    while True:
        print("""
1. View Profile
2. View Marks & Result
3. Logout
""")
        ch = input("Choice: ")

        if ch == "1":
            print("----- Profile -----")
            print("Roll:", student["roll"])
            print("Name:", student["name"])
            print("Age:", student["age"])
            print("Course:", student["course"])

        elif ch == "2":
            print("Marks:", marks)
            print("Total:", total)
            print("Percentage:", percentage)
            print("Grade:", grade)

        elif ch == "3":
            break
        else:
            print("Invalid Choice")
```

## üñ•Ô∏è Main Menu

This is the entry point of the system.

```python
def admin_menu():
    while True:
        print("""
1. Add Student
2. View Students
3. Update Student
4. Delete Student
5. Logout
""")
        ch = input("Choice: ")

        if ch == "1":
            add_student()
        elif ch == "2":
            view_students()
        elif ch == "3":
            update_student()
        elif ch == "4":
            delete_student()
        elif ch == "5":
            break
        else:
            print("Invalid Choice")


def main():
    while True:
        print("""
====== Student Management System ======
1. Admin Login
2. Student Login
3. Exit
""")
        choice = input("Enter Choice: ")

        if choice == "1":
            admin_login()
        elif choice == "2":
            student_login()
        elif choice == "3":
            print("Program Exit")
            break
        else:
            print("Invalid Choice")


main()
```
